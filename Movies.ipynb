```python
import pandas as pd
import numpy as np
import ast
import string
```

```python
url = "https://raw.githubusercontent.com/nnqomariyah/Fundamentals_of_Data_Science/main/week_3/data/movies_metadata.csv"
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df['belongs_to_collection'].fillna("{}", inplace=True)
df['genres'].fillna("[]", inplace=True)
df['production_companies'].fillna("[]", inplace=True)
df['production_countries'].fillna("[]", inplace=True)
df['spoken_languages'].fillna("[]", inplace=True)
df['overview'].fillna("", inplace=True)
df['budget'].fillna(0, inplace=True)
df['revenue'].fillna(0, inplace=True)
df['runtime'].fillna(df['runtime'].mean(), inplace=True)

print(df.isnull().sum())

df.fillna("Unknown", inplace=True) 
```

```python
print(df.isnull().sum())
```

```python
def parse_json(column, key):
    try:
        result = "|".join([d[key] for d in ast.literal_eval(column)])
    except (SyntaxError, ValueError, KeyError):
        result = ""
    return result

json_columns = ['belongs_to_collection', 'genres', 'production_companies', 'production_countries', 'spoken_languages']
```

```python
def clean_text(text):
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

df['clean_overview'] = df['overview'].apply(clean_text)
```

```python
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')

# Convert 'release_date' to datetime
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
print(df.duplicated().sum())
df = df.drop_duplicates()
```
