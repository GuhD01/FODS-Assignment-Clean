```python
df_merge.rename(columns={
    'season_x': 'match_season',
    'venue_x': 'match_venue',
    'venue_y': 'info_venue',
}, inplace=True)

print(df_merge.columns)
```

```python
df_merge.set_index('match_id', inplace=True)

df_merge.index.name = 'Match_ID'

df_merge.index = df_merge.index + 1000

df_merge.reset_index(drop=False, inplace=True)

print(df_merge.head())
```

```python
pd.set_option("display.max_rows", 10)  
```

```python
# Melting the dataframe using the index
melted_df = df_merge.melt(id_vars=['Match_ID', 'striker'], 
                          value_vars=['runs_off_bat', 'extras'], 
                          var_name='run_type', 
                          value_name='run_value')

print(melted_df.head())
```

```python
# Pivoting the dataframe
pivoted_df = melted_df.pivot_table(index=['Match_ID', 'striker'], 
                                   columns='run_type', 
                                   values='run_value', 
                                   aggfunc='sum').reset_index()

print(pivoted_df.head())
```

```python
# Using df.apply() to get the maximum value of each numeric column
max_values = pivoted_df.apply(max, axis=0)
print("Max values of each column:")
print(max_values)

# Filter numeric columns
numeric_cols = pivoted_df.select_dtypes(include=[np.number])

# Using df.apply() to get the sum of each row for numeric columns
row_sums = numeric_cols.apply(sum, axis=1)
print("\nSum of each row for numeric columns:")
print(row_sums.head())

# Using df.applymap() to convert the entire dataframe into string format
string_df = pivoted_df.applymap(str)
print("\nDataframe in string format:")
print(string_df.head())
```

```python
# Fill NaN in numeric columns with their mean
numeric_cols = ODI_Match_Data.select_dtypes(include=[np.number])
numeric_filled = numeric_cols.fillna(numeric_cols.mean())
non_numeric_cols = ODI_Match_Data.select_dtypes(exclude=[np.number])
non_numeric_filled = non_numeric_cols.fillna('Unknown')
filled_data = pd.concat([numeric_filled, non_numeric_filled], axis=1)

print("Missing values before handling:")
print(ODI_Match_Data.isnull().sum())

print("\nMissing values after handling:")
print(filled_data.isnull().sum())
```

```python
import re

# Function to clean text data using regex
def clean_text(text):
    text = str(text).lower()  
    text = re.sub(r'[^a-z\s]', '', text)  
    text = text.strip()  
    return text
df_merge['striker'] = df_merge['striker'].apply(clean_text)
print(df_merge['striker'].head())
```
